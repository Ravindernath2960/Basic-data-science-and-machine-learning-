```python

import numpy as np
i = data[data['BloodPressure'] == 0].index
data = data.drop(i)
i
```

```python
data['Outcome'] = data['Outcome'].replace(np.NaN,0) 
data
```

```python
data.isna().sum()
```

```python
for i in data.columns:
        q1 = data[i].quantile(0.25)
        q3 = data[i].quantile(0.75)
        iqr = q3 - q1
        lower_limit = q1 - 1.5*iqr
        upper_limit = q3 + 1.5*iqr
        data = data[(data[i] > lower_limit)&(data[i] < upper_limit)]
```
