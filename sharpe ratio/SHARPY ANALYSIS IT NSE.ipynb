```python
mean_pct_change = excess_change.mean()
sd_pct_change = excess_change.std()
```

```python
mean_pct_change
```

```python
sd_pct_change
```

```python
annual_factor = np.sqrt(252)
sharpe_ratio = mean_pct_change.mul(annual_factor).div(sd_pct_change)
```

```python
sharpe_ratio.plot.bar(color = ['red', 'blue', 'green', 'yellow', 'orange', 'black', 'brown'])
plt.title("SHARPY RATIO NIFTY IT STOCK")
plt.savefig("sharpy_nifty_it.jpeg")
plt.show()
```
